```python
loan_application_test = pd.read_csv('DataSet/loan_applications_test.csv')
previous_loan_applications = pd.read_csv('DataSet/previous_loan_applications.csv')

merged_loan_data = pd.merge(loan_application_test, previous_loan_applications, on='sk_id_curr', how='left')
```

```python
previous_credits = pd.read_csv('DataSet/previous_credits.csv')

merged_loan_data.merge(previous_credits, on='sk_id_curr', how='left')
```

```python
previous_credit_cards = pd.read_csv('DataSet/previous_credit_cards.csv')

merged_loan_data.merge(previous_credit_cards, on='sk_id_curr', how='left')
```

```python
previous_pos_cash_loans = pd.read_csv('DataSet/previous_pos_cash_loans.csv')

merged_loan_data.merge(previous_pos_cash_loans, on='sk_id_curr', how='left')
```

```python
repayment_history = pd.read_csv('DataSet/repayment_history.csv')

merged_loan_data.merge(repayment_history, on='sk_id_curr', how='left')
```

```python
credit_bureau_balance = pd.read_csv('DataSet/credit_bureau_balance.csv')

merged_loan_data.merge(credit_bureau_balance, on='sk_id_curr', how='left')
```
